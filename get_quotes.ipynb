```python
url = "http://www.imdb.com/character/ch0146096/quotes"
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content)
```

```python
import re
import json

text = soup.select('#tn15content')[0].text

reg = re.compile('([A-Z][a-z]+ [A-Z][a-z]+:)')

blocks = [block.strip(' \n:').replace('\n', '') for block in reg.split(text)]

import numpy as np

b = np.array(blocks)

quotes = b[np.where([b == "Tyrion Lannister"])[1] + 1].tolist()
with open('quotes.txt', 'w') as f:
    json.dump(quotes, f)
```
